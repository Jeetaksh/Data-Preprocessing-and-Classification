```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a Pandas DataFrame
df = pd.read_csv('ad.csv')

# Define the two classes based on the values in the 15th column (assuming income is the 15th column)
class1 = [">50K", ">50K."]
class2 = ["<=50K", "<=50K."]

# Get unique education levels from the entire DataFrame
unique_education_levels = df['education'].unique()

# Filter the DataFrame for each class and count the occurrences of each education level
class1_counts = df[df.iloc[:, 14].str.strip().isin(class1)]['education'].value_counts().reindex(unique_education_levels).fillna(0)
class2_counts = df[df.iloc[:, 14].str.strip().isin(class2)]['education'].value_counts().reindex(unique_education_levels).fillna(0)

# Combine the education levels and their corresponding counts into a DataFrame
result_df = pd.DataFrame({
    'Education': unique_education_levels,
    'Count_Class1': class1_counts.tolist(),
    'Count_Class2': class2_counts.tolist()
})

# Plot the data
plt.figure(figsize=(12, 6))
plt.bar(range(len(result_df)), result_df['Count_Class1'], width=0.4, label='>50K or >50K.')
plt.bar([x + 0.4 for x in range(len(result_df))], result_df['Count_Class2'], width=0.4, label='<=50K or <=50K.')

# Set x-axis ticks and labels
plt.xticks(range(len(result_df)), result_df['Education'], rotation=45, ha="right")

plt.xlabel('Education Level')
plt.ylabel('Number of People')
plt.title('Number of People by Education Level and Income Class')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a pandas DataFrame
df = pd.read_csv('ad.csv')

# Count the occurrences of each age
age_counts = df['education'].value_counts()
print(age_counts)

# Plot a pie chart with plain color circle
plt.figure(figsize=(10, 6))
plt.pie(age_counts, labels=None, autopct='', startangle=90)  # No labels inside the pie chart
plt.gca().set_aspect('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle.

# Display percentages in the legend
percentages = ['{:.1f}%'.format(x) for x in (age_counts / age_counts.sum() * 100)]
labels_with_percentages = [f'{label} ({percentage})' for label, percentage in zip(age_counts.index, percentages)]
print(percentages)

plt.legend(labels_with_percentages, title='Education', loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('Education Distribution')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a Pandas DataFrame
df = pd.read_csv('ad.csv')

# Combine categories '<=50K' and '<=50K.' into one category
df['income_0'] = df['income_0'].replace({' <=50K.': ' <=50K'})

# Combine categories '>50K' and '>50K.' into one category
df['income_0'] = df['income_0'].replace({' >50K.': ' >50K'})

# Filter the DataFrame based on specified conditions
filtered_df = df[(df['occupation'] == ' Exec-managerial') & (df['education'] == ' Masters') & (df['MARITAL_STATUS'] == ' Married-civ-spouse')]

# Count the occurrences of each category in the 'income_0' field
income_counts = filtered_df['income_0'].value_counts()

# Plot a pie chart
plt.figure(figsize=(8, 8))
plt.pie(income_counts, labels=income_counts.index, autopct='%1.1f%%', startangle=90)
plt.gca().set_aspect('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle.

plt.title('Distribution of Exec-Managerial, Married, Masters by Income')
plt.show()
```
